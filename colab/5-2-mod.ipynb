## 검증 세트

```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
```

```python
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42)
```

```python
print(sub_input.shape, val_input.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)

print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
# 전체 데이터셋
# ┌─────────────────────────────┐
# │                      전체 데이터 (100%)                  │
# └─────────────────────────────┘
#                          ↓
# 1단계: Train / Test 분할 (예: 80% / 20%)
# ┌───────────────┐ ┌────────────┐
# │         Train (80%)          │ │ Test (20%)             │
# └───────────────┘ └────────────┘
#                          ↓
# 2단계: Train → Sub-train / Validation 분할 (예: 64% / 16%)
# ┌───────┐ ┌───────┐
# │ Sub-train    │ │ Validation   │
# │ (64%)        │ │ (16%)        │
# └───────┘ └───────┘
```

## 교차 검증

```python
from sklearn.model_selection import cross_validate
# dt = DecisionTreeClassifier(random_state=42)
scores = cross_validate(dt, train_input, train_target)
print(scores)
```

```python
import numpy as np

print(np.mean(scores['test_score']))
```

```python
from sklearn.model_selection import StratifiedKFold

scores = cross_validate(dt, train_input, train_target, cv=StratifiedKFold())
print(np.mean(scores['test_score']))
```

```python
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_validate(dt, train_input, train_target, cv=splitter, return_train_score=True)
print(scores)
print(np.mean(scores['test_score']))
```

## 하이퍼파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]}
```

```python
gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
```

```python
gs.fit(train_input, train_target)
```

```python
import matplotlib.pyplot as plt

results = gs.cv_results_
plt.figure(figsize=(10, 5))
plt.plot(params['min_impurity_decrease'], results['mean_test_score'])
plt.xlabel('min_impurity_decrease')
plt.ylabel('Mean Test Score')
plt.title('Grid Search Results')
```

```python
dt = gs.best_estimator_
print(dt.score(train_input, train_target))
```

```python
print(gs.best_params_)
```

```python
print(gs.cv_results_['mean_test_score'])
```

```python
best_index = np.argmax(gs.cv_results_['mean_test_score'])
print(gs.cv_results_['params'][best_index])
```

```python
params = {'min_impurity_decrease': np.arange(0.0001, 0.001, 0.0001),
          'max_depth': range(5, 20, 1),
          'min_samples_split': range(2, 100, 10)
          }
```

```python
gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
gs.fit(train_input, train_target)
```

```python
print(gs.best_params_)
```

```python
print(np.max(gs.cv_results_['mean_test_score']))
```

### 랜덤 서치

```python
from scipy.stats import uniform, randint
```

```python
np.unique(rgen.rvs(1000), return_counts=True)
```

```python
ugen = uniform(0, 1)
ugen.rvs(10)
```

```python
params = {'min_impurity_decrease': uniform(0.0001, 0.001),
          'max_depth': randint(20, 50),
          'min_samples_split': randint(2, 25),
          'min_samples_leaf': randint(1, 25),
          }
```

```python
from sklearn.model_selection import RandomizedSearchCV

gs = RandomizedSearchCV(DecisionTreeClassifier(random_state=42), params,
                        n_iter=100, n_jobs=-1, random_state=42)
gs.fit(train_input, train_target)
```

```python
print(gs.best_params_)
```

```python
print(np.max(gs.cv_results_['mean_test_score']))
```

```python
dt = gs.best_estimator_

print(dt.score(test_input, test_target))
```

```python
import seaborn as sns

# 결과를 DataFrame으로 변환
results = pd.DataFrame(gs.cv_results_)

# print(results.head())

# 산점도 시각화 (max_depth vs min_samples_split vs 성능)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=results,
    x='param_max_depth',
    y='param_min_samples_split',
    hue='mean_test_score',
    size='mean_test_score',
    palette='viridis',
    sizes=(40, 200),
    legend='brief'
)

plt.title('RandomizedSearchCV: max_depth vs min_samples_split')
plt.xlabel('max_depth')
plt.ylabel('min_samples_split')
plt.grid(True)
# plt.colorbar(scatter_mappable.collections[0], label='mean_test_score') # scatterplot returns multiple collections, the first one is the one we need for colorbar
# plt.colorbar(gs.cv_results_['mean_test_score'], label='mean_test_score')
plt.show()
```

```python
# 주요 파라미터들만 선택
plot_data = results[[
    'param_max_depth',
    'param_min_samples_split',
    'param_min_samples_leaf',
    'param_min_impurity_decrease',
    'mean_test_score'
]]

# pairplot으로 다중 변수 관계 시각화
sns.pairplot(plot_data, corner=True, diag_kind="kde", hue="mean_test_score", palette="viridis")
plt.suptitle("Random Search Results", y=1.02)
plt.show()
```
